```python
# Import modul yang dibutuhkan
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, accuracy_score
```

```python
# Membaca data dari file Excel
data = pd.read_excel('/content/drive/MyDrive/Colab Notebooks/_Tugas_Akhir/data/data_panel.xlsx')
```

```python
# Menambahkan normalisasi data p
numeric_columns = data.select_dtypes(include=[np.number]).columns
data[numeric_columns] += np.random.normal(scale=0.0001, size=data[numeric_columns].shape)
data['Throughput_DL_rata_rata_pengguna'] += np.random.normal(scale=0.0001, size=len(data['Throughput_DL_rata_rata_pengguna']))
X = data[numeric_columns]
y_pengguna = data['Throughput_DL_rata_rata_pengguna']
```

```python
# Membagi data menjadi set pelatihan dan pengujian
X_train, X_test, y_train, y_test = train_test_split(X, y_pengguna, test_size=0.2, random_state=32)
```

```python
# Membuat model MLR
model_mlr_pengguna = LinearRegression()
model_mlr_pengguna.fit(X_train, y_train)
```

```python
# Melakukan prediksi
y_pred_pengguna = model_mlr_pengguna.predict(X_test)
value_peng = max(y_pred_pengguna.max(), y_test.max())
thresholds = value_peng - 0
y_tes_peng_binary = np.where(y_pred_pengguna >= thresholds, 1, 0)
y_pred_peng_binary = np.where(y_test >= thresholds, 1, 0)
```

```python
# Menghitung metrik evaluasi
mse_pengguna = mean_squared_error(y_test, y_pred_pengguna)
mae_pengguna = mean_absolute_error(y_test, y_pred_pengguna)
accuracy_pengguna = accuracy_score(y_tes_peng_binary, y_pred_peng_binary)
```

```python
# Tampilkan Metrics akurasi
print("\nMLR Metrics - Throughput Pengguna:")
print(f"1. Akurasi: {accuracy_pengguna}")
print(f"2. Mean Squared Error (MSE): {mse_pengguna}")
print(f"3. Mean Absolute Error (MAE): {mae_pengguna}")
```

# THROUGHPUT PER SEL

```python
# Memisahkan variabel dependen untuk throughput per sel
y_sel = data['Throughput_DL_rata_rata_sel']
```

```python
# Membagi data menjadi set pelatihan dan pengujian
X_train_sel, X_test_sel, y_sel_train, y_sel_test = train_test_split(X, y_sel, test_size=0.2, random_state=32)
```

```python
# Membuat model MLR untuk throughput per sel
model_mlr_sel = LinearRegression()
model_mlr_sel.fit(X_train_sel, y_sel_train)
```

```python
# Melakukan prediksi
y_pred_sel = model_mlr_sel.predict(X_test_sel)
value_sel = max(y_pred_sel.max(), y_sel_test.max())
threshold = value_sel - 0
y_tes_sel_binary = np.where(y_pred_sel >= threshold, 1, 0)
y_pred_sel_binary = np.where(y_sel_test >= threshold, 1, 0)
```

```python
# Menghitung metrik evaluasi
mse_sel = mean_squared_error(y_sel_test, y_pred_sel)
mae_sel = mean_absolute_error(y_sel_test, y_pred_sel)
accuracy = accuracy_score(y_tes_sel_binary, y_pred_sel_binary)
```

```python
# Mencetak Metrics
print("\nMLR Metrics - Throughput per Sel:")
print(f"1. Akurasi: {accuracy}")
print(f"2. Mean Squared Error (MSE): {mse_sel}")
print(f"3. Mean Absolute Error (MAE): {mae_sel}")
```
